### b. Pemeriksaan Local Outlier Factor(LOF)

Langkah selanjutnya adalah deteksi outlier. Deteksi outlier adalah metode untuk mengidentifikasi titik-titik data yang dianggap sebagai outlier atau anomali. Outlier adalah nilai yang secara signifikan berbeda dari sebagian besar data lainnya dalam sebuah dataset. Mengidentifikasi outliers bermanfaat untuk meningkatkan akurasi analisis dan model prediktif.

```
# Membuat Encode Sex
from sklearn.preprocessing import LabelEncoder

#inisilalisasi objek LabelEncoder
label_encoder = LabelEncoder()

# melakukan label encoding pada fitur 'Sex'
df['Sex'] = label_encoder.fit_transform(df['Sex'])

# menampilkan mapping antara nilai asli dengan nilai  yang terekripsi

label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Mapping Label Encoding:", label_mapping)

df
```

```
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
import numpy as np


# Memastikan kolom 'Sex' ada di dataset
if 'Sex' not in df.columns:
    raise ValueError("Kolom 'Sex' tidak ditemukan dalam dataset")

# Siapkan data untuk LOF (drop kolom 'Sex' karena fitur target)
x = df.drop('Sex', axis=1)

# Inisialisasi model LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.036)

# Terapkan LOF pada dataset
y_pred = lof.fit_predict(x)  # Hasil prediksi berupa label -1 untuk outlier dan 1 untuk inlier.
# Skor anomali (semakin rendah, semakin besar kemungkinan outlier)
anomaly_scores = lof.negative_outlier_factor_  # Skor anomali dari LOF

# Tambahkan kolom hasil prediksi dan skor anomali ke dalam dataframe
df['LOF_Prediksi'] = y_pred  # Hasil prediksi LOF
df['LOF_Skor_Anomali'] = anomaly_scores  # Skor anomali dari LOF

# Visualisasi hasil
plt.figure(figsize=(10, 6))  # Membuat figure baru dengan ukuran 10x6
plt.title("LOF Anomaly Scores")  # Memberi judul pada plot

# Definisikan warna untuk scatter plot berdasarkan target 'Sex'
# colors = df['Sex'].apply(lambda x: 'pink' if x == 'Male' else 'green')

# Scatter plot dari skor anomali dengan warna berdasarkan target 'Sex'
plt.scatter(np.arange(len(anomaly_scores)), anomaly_scores, c=(y_pred==-1), cmap='coolwarm',
            edgecolor='k', label=df['LOF_Prediksi'])
plt.axhline(y=-2, color='r', linestyle='--')  # Garis horizontal pada y = -1.5 untuk memisahkan outlier
plt.xlabel("Indeks")
plt.ylabel("Skor Anomali LOF")
plt.colorbar(label='Sex')  # Menambahkan colorbar dengan label 'Sex'
plt.show()

# Print beberapa baris pertama untuk melihat hasil
print(df.head())
print(y_pred)
```

### -KNN

Proses algoritma KNN dapat dijelaskan dalam beberapa tahap:

1. Menentukan parameter K: Parameter K adalah jumlah tetangga terdekat yang akan digunakan dalam proses prediksi. Nilai K harus ditentukan sebelumnya oleh pengguna.

2. Menghitung Jarak: Algoritma menghitung jarak antara titik data yang akan diprediksi dengan setiap titik data dalam dataset pelatihan. Jarak ini biasanya dihitung menggunakan metrik jarak seperti Euclidean distance, Manhattan distance, atau Minkowski distance.

3. Memilih Tetangga Terdekat: Algoritma memilih K titik data dengan jarak terdekat dengan titik data yang akan diprediksi.

4. Menghitung Prediksi: Untuk masalah klasifikasi, algoritma mengambil mayoritas dari label-label tetangga terdekat sebagai prediksi kelas untuk titik data yang tidak diketahui. Untuk regresi, algoritma mengambil rata-rata dari nilai target tetangga terdekat sebagai prediksi.

Mencari akurasi data menggunakan metode KNN

MELAKUKAN PREDIKSI AKURASI DENGAN BAGGING

1. Pertama fungsi akan dibuat dengan nama bagging dan mengambil parameter x_train, y_train,x_test,y_test,jumlah_juri
2. Setelah itu melakukan perulangan yang dimulai dari i = 3 sampai jumlah_juri

Pada perulangan fungsi BaggingClassifier dari sklearn akan dipanggil dengan menginputkan beberapa paramter Yang pertama base_estimator berfungsi untuk memberitahu metode apa yang ingin kita gunakan kali ini kita akan memakai GaussianNB(), n_estimator berfungsi untuk memberitahu berapa banyak juri yang akan kita gunakan, pada paramater ini inputnya dinamis mengikuti nilai dari i.
4. Selanjutnya fit (latih) dengan x_trian dan y_train dari inputan parameter kita sebelumnya
5. Setelah itu prediksi x_test dengan model yang kita latih sebelumnya sesuai dengan n_estimator
6. Lalu hasil dari prediksi kita Encoder ke tipe numeric agar bisa kita hitung akurasinya
7. Selanjutnya memasukkan data ke array accuracy yang kita akan gunakan sebagai pengelompokan accuracy dari beberapa n_estimator
8.Selanjutnya akan memilih tingkat akurasi tertinggi dan jumlah n_estimator terendah
pada percobaan kali ini kita menemukan akurasi tertinggi sebesar 52.72% dengan n_estimator = 3
